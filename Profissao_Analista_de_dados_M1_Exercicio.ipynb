## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
print("Valor total de vendas")
qtd_total_vendas = int(3)
ticket_medio = float(320.52)
valor_total_vendas = float(qtd_total_vendas * ticket_medio)
print("O valor de A será: ", valor_total_vendas)
```

```
print("Quantidade Total de Vendas")
valor_total_vendas = float(834.47)
ticket_medio = float(119.21)
qtd_total_vendas = int(valor_total_vendas / ticket_medio)
print("O valor de B será: ", qtd_total_vendas)
```

```
print("Ticket Médio")
valor_total_vendas = float(15378.12)
qtd_total_vendas = int(5)
ticket_medio = float(valor_total_vendas / qtd_total_vendas)
C = round(ticket_medio, 2)
print("O valor de C será: ", C)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

print("#" * 40)
print("*Numero de letras O na frase*")
contador = cancao.count("o")
print(contador)
print("#" * 40)
print("\n")

print("#" * 40)
print("*Substituindo a palavra 'gigante'*")
troca = cancao.replace("gigante", "anã")
print(troca)
print("#" * 40)
print("\n")

print("#" * 40)
print("*Torndo a frase maiúscula*")
capital = cancao.upper()
print(capital)
print("#" * 40)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

posicao = noticia.find('2')
selic = noticia[posicao:(posicao+5)]
showsel = f'Selic = {selic}'
print(showsel)
find6 = noticia.find('6')
ano = noticia[find6]
showano = f'Ano = {ano}'
print(showano)
```
